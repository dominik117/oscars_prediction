The following code upload the stage3 datafram to a MariaDB databse

```python
import mariadb
import pandas as pd
import sqlalchemy
```

## Make connection

```python
# Make the connection to MariaDB
user="cip_user"
password="cip_pw"
host="localhost" #"127.0.0.1", 
port="3306"
database="OSCARS_STUDENTA"

engine = sqlalchemy.create_engine(f"mariadb+mariadbconnector://{user}:{password}@{host}:{port}/{database}")
```

## Upload Data to MariaDB

```python
# Open the original CSV files
df_stage3 = pd.read_csv("../Data/Bacher_Dominik_studentA_stage3.csv")
```

```python
df_stage3.head(3)
```

```python
# Create the table in MariaDB from original CSV
df_stage3.to_sql(
    'Bacher_Dominik_studentA_stage3', 
    con=engine, 
    index=False, 
    if_exists='replace'
)
```

## Load from MariaDB

```python
df_stage3_mariadb = pd.read_sql_table('Bacher_Dominik_studentA_stage3', con=engine)

print(F"Loaded {len(df_stage3_mariadb)} rows from the table 'Bacher_Dominik_studentA_stage3' from MariaDB")
print("Here is a sample:")
df_stage3_mariadb
```

```python
# Inspect the tables inside the database
inspector = sqlalchemy.inspect(engine)

tables = inspector.get_table_names()
print("List of tables in the database 'OSCARS_STUDENTA':")
for table in tables:
    print(table)
```

```python
# Code to drop tables:
drop_statement = sqlalchemy.text("DROP TABLE IF EXISTS table3;")

with engine.connect() as connection:
    connection.execute(drop_statement)
    connection.commit() 
```
